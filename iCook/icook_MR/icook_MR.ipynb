```python
# get each ingredient without its amount
def ingredientList(sb):
    list = []
    for i in (sb.replace('\n', ',').split(',')[5::4]):
        for m, n in enumerate(i.replace('}{', ':').replace('{', '').replace('}', '').split(':')):
            if m%2 == 0:
                list.append(n)
    return list

# merge multiple lists into one list
def mergeAllList(list, *args):
    for i in args:
        for j in i:
            list.append(j)
    return list

# reduce -> compute amount of each ingredient
def reduceListToDict(list):
    ingredientDict = {}
    for i in list:
        if i in ingredientDict:
            ingredientDict[i] += 1
        else:
            ingredientDict[i] = 1
    return ingredientDict
```

```python
list = []
for i in os.listdir(r'./res'):
    for j in os.listdir(r'./res/%s'%i):
        sb = ''
        with open(r'./res/%s/%s'%(i, j), 'r', encoding='utf-8') as f:
            sb = f.read()
        mergeAllList(list, ingredientList(sb))
ingredientAmountDict = reduceListToDict(list)
ingredientAmountDict
```

```python
'''

Remove all dirty datas
Transform all csv to dataframe

'''

import pandas as pd
import os

path = r'./res_clean'
if not os.path.isdir(path):
    os.mkdir(path)
    
# initiate dataFrame
df = pd.DataFrame(columns = ['Cuisine', 'Ingredients', 'image', "Url"])

# open each directory
for i in os.listdir(r'./res'):
    print('[%s]'%i)
    # open each csv
    rowData = []
    for j in os.listdir(r'./res/%s'%i):
        
        print(r'./res/%s/%s'%(i, j))
        
        # split by "\n", and then split again by ","
        with open(r'./res/%s/%s'%(i, j), 'r', encoding = 'utf-8') as f:
            rowData = f.read().split('\n')    
        rowData.pop(0)
        for k in rowData:
            # check each row data, if len(eachRowData) != 4 , delete it(do not save it)
            # if any block cantain nothing, delete it(do not save it)
            if (len(k.split(',')) == 4) and ('' not in k.split(',')):
                df = df.append(
                    pd.DataFrame(
                        [k.split(',')],
                        columns = ['Cuisine', 'Ingredients', 'image', "Url"]
                    ), 
                    ignore_index = True
                )

        print('Done.')
    print('-')

df.to_csv(r'./res_clean/iCook_clean.csv', sep=',', index=0) 
print('All done.')
```

```python
'''

From res_clean
Filter ingredients we need

'''

import pandas as pd
import os

path = r'./res_filter'
if not os.path.isdir(path):
    os.mkdir(path)
    
# initiate dataFrame
df = pd.DataFrame(columns = ['Cuisine', 'Ingredients', 'image', "Url"])

def df_append(df, dt):
    df = df.append(
        pd.DataFrame(
            [dt],
            columns = ['Cuisine', 'Ingredients', 'image', "Url"]
        ), 
        ignore_index = True
    )
    return df

# open each directory
for i in os.listdir(r'./res'):
    print('[%s]'%i)
    # open each csv
    rowData = []
    for j in os.listdir(r'./res/%s'%i):
        
        print(r'./res/%s/%s'%(i, j))
        
        # split by "\n", and then split again by ","
        with open(r'./res/%s/%s'%(i, j), 'r', encoding = 'utf-8') as f:
            rowData = f.read().split('\n')    
        rowData.pop(0)
        for k in rowData:
            # check each row data, if len(eachRowData) != 4 , delete it(do not save it)
            # if any block cantain nothing, delete it(do not save it)
            if (len(k.split(',')) == 4) and ('' not in k.split(',')):                
                # check [鹽 醬油 蛋 麵粉 奶油 洋蔥 蔥 肉]
                check_ingredient_used = 0                
                if '鹽' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '醬油' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '蛋' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '麵粉' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '奶油' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '洋蔥' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '蔥' in k.split(',')[1]:
                    check_ingredient_used += 1
                if '肉' in k.split(',')[1]:
                    if '雞' in k.split(',')[1]:
                        check_ingredient_used += 1
                    elif '豬' in k.split(',')[1]:
                        check_ingredient_used += 1
                    elif '牛' in k.split(',')[1]:
                        check_ingredient_used += 1
                    elif '羊' in k.split(',')[1]:
                        check_ingredient_used += 1
                    elif '魚' in k.split(',')[1]:
                        check_ingredient_used += 1
                if check_ingredient_used > 1:
                    df = df_append(df, k.split(','))
        print('Done.')
    print('-')

df.to_csv(r'%s/iCook_filter.csv'%(path), sep=',', index=0) 
print('All done.')
```
