```python
'''

Load one-hot encoding data

'''
import pandas as pd

path = r'./res_classified'

oneHotEncoding = pd.read_csv(r'%s/iCook_ingredient_one_hot_encoding_5.csv'%(path), engine='python', encoding='utf-8')
oneHotEncoding.drop(['rdIndex'], axis=1)
```

```python
'''

Compare various N 2~100 in Kmeans

'''

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

data_used = oneHotEncoding.drop(['rdIndex'], axis=1)

axis_x = []
axis_y = []

for k in range(2,100):
    print('[k=%s]\t'%k, '==', end='')
    clf = KMeans(n_clusters = k)
    print('==', end='')
    clf.fit(data_used)
    print('==', end='')
    axis_x.append(k)
    print('==', end='')
    axis_y.append(silhouette_score(data_used, clf.labels_))
    print('== -> Done.')

print('----------')
for n, i in enumerate(axis_x):
    print('k=%s  -> '%i, '%s'%axis_y[n])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(axis_x, axis_y, 'm^--')
plt.title('Sihouette Score')
plt.xlabel('K')
plt.ylabel('Score')
```
