# Data cleaning
## 8 ingredients

```python
import pandas as pd
import os

path = r'./res_analyze_used'

df_origin = pd.read_csv(r'%s/iCook_ingredient_res_analyze_used_v5.csv'%(path), engine='python', encoding='utf-8')
df = df_origin
df_origin
```

```python
'''

é¹½ã€é†¬æ²¹ã€è›‹ã€éºµç²‰ã€å¥¶æ²¹ã€æ´‹è”¥ã€è”¥ã€è‚‰
salt, soy sauce, egg, flour, butter, onion, shallot, meat(pork, beef, fish, chicken), water

'''

ingredient_row_data = df_origin['Ingredients']
ingredient_row_data
```

```python
'''

Compute the amount of all ingredient
Map-Reduce

'''

# Map-Reduce
each_ingredient_amount_dict = {}

for num, each_row in enumerate(ingredient_row_data):
    try:
        for each_ingredient in each_row.split(';'):
            if each_ingredient in each_ingredient_amount_dict:
                each_ingredient_amount_dict[each_ingredient] += 1
            else:
                each_ingredient_amount_dict[each_ingredient] = 1
    except AttributeError:
        print(num, 'AttributeError')
        #ingredient_row_data.drop([num])
        continue

len(each_ingredient_amount_dict)
```

```python
adv_each_ingredient_amount_dict = {}
for i in each_ingredient_amount_dict:
    if each_ingredient_amount_dict[i] > 7500:
        adv_each_ingredient_amount_dict[i] = each_ingredient_amount_dict[i]
len(adv_each_ingredient_amount_dict)
```

```python
adv_each_ingredient_amount_dict
```

```python
necessary_ingredient_list = [i for i in adv_each_ingredient_amount_dict]
necessary_ingredient_list
```

```python
necessary_ingredient_list += ['pork', 'beef', 'fish', 'chicken', 'sheep', 'lamb']
necessary_ingredient_list
```

```python
'''

Drop those cuisines which are not composed with ['pepper',
 'egg',
 'water',
 'salt',
 'oil',
 'soy sauce',
 'onion',
 'sugar',
 'cream',
 'flour',
 'pork',
 'beef',
 'fish',
 'chicken',
 'sheep',
 'lamb']

'''
drop_list = []
for num, each_row in enumerate(ingredient_row_data):
    check_n = 0
    try:
        for each_ingredient in each_row.split(';'):
            if not each_ingredient in necessary_ingredient_list:
                check_n = 1
        
    except:
        check_n = 1

    if check_n == 1:
        drop_list.append(num)
        ingredient_row_data.drop([num])
    else:
        print(str(num)+',', end='')
        
print('\n-------------------\nOrigin:%s , Drop:%s , Left:%s'%(len(ingredient_row_data), len(drop_list), len(ingredient_row_data)-len(drop_list)))
```

```python
ingredient_row_data
```

```python
df = df_origin
df = df.drop(drop_list)
df
```

```python
len(df)
```

```python
path = r'./res_second_analyze'
if not os.path.isdir(path):
    os.mkdir(path)

df.to_csv(r'%s/iCook_ingredient_res_analyze_used_v5_8_left_v1.csv'%(path), sep=',', index=0)
```

# Kmeans

```python
path = r'./res_second_analyze'
if not os.path.isdir(path):
    os.mkdir(path)

df_origin = pd.read_csv(r'%s/iCook_ingredient_res_analyze_used_v5_8_left_v1.csv'%(path), engine='python', encoding='utf-8')
df = df_origin
df_origin
```

```python
ingredient_row_data = df_origin['Ingredients']
ingredient_row_data
```

```python
'''

According to the result, do normalization

'''

adv_each_ingredient_amount_list = [i for i in adv_each_ingredient_amount_dict]
adv_each_ingredient_amount_list += ['pork', 'beef', 'fish', 'chicken', 'sheep', 'lamb', 'meat']

col = ['rdIndex'] + [ingredient for ingredient in adv_each_ingredient_amount_dict] + ['pork', 'beef', 'fish', 'chicken', 'sheep', 'lamb', 'meat']

nor_df = pd.DataFrame(columns=col)

# Generate empty table
print('Generate empty table...')
for num, single_ingredient_row_data in enumerate(ingredient_row_data):
    print('[%s]'%num, end='')
    nor_df = nor_df.append(pd.DataFrame([[num]+[0 for i in adv_each_ingredient_amount_list]], columns=col), ignore_index=True)
        
nor_df
```

```python
'''

One-hot encoding


'''

# Generate normalized table
errorList = []
print('Generate normalized table...')
for num, single_ingredient_row_data in enumerate(ingredient_row_data):
    print('[%s]'%num)
    nor_df['rdIndex'][num] = num
    try:
        for single_ingredient in single_ingredient_row_data.split(';'):
            if single_ingredient in adv_each_ingredient_amount_list:
                nor_df[single_ingredient][num] = 1
        for meat in ['pork', 'beef', 'fish', 'chicken', 'sheep', 'lamb']:
            if meat in single_ingredient_row_data.split(';'):
                nor_df['meat'][num] = 1
                
    except AttributeError:
        print('AttributeError')
        errorList.append(num)
    print(col)
    print(nor_df[num:num+1])
```

```python
'''

Load one-hot encoding data

'''
import pandas as pd

path = r'./res_second_analyze'

oneHotEncoding = pd.read_csv(r'%s/iCook_ingredient_one_hot_encoding_5_8_left.csv'%(path), engine='python', encoding='utf-8')
oneHotEncoding.drop(['rdIndex'], axis=1)
```

```python
'''

Compare various N 2~20 in Kmeans

'''

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

data_used = oneHotEncoding.drop(['rdIndex', 'pork', 'beef', 'fish', 'chicken', 'sheep', 'lamb', 'water'], axis=1)

axis_x = []
axis_y = []

for k in range(2,21):
    print('[k=%s]\t'%k, '==', end='')
    clf = KMeans(n_clusters = k)
    print('==', end='')
    clf.fit(data_used)
    print('==', end='')
    axis_x.append(k)
    print('==', end='')
    axis_y.append(silhouette_score(data_used, clf.labels_))
    print('== -> Done.')

print('----------')
for n, i in enumerate(axis_x):
    print('k=%s  -> '%i, '%s'%axis_y[n])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(axis_x, axis_y, 'm^--')
plt.title('Sihouette Score')
plt.xlabel('K')
plt.ylabel('Score')
```

```python
'''

Compare various N 21~100 in Kmeans

'''

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

data_used = oneHotEncoding.drop(['rdIndex', 'pork', 'beef', 'fish', 'water'], axis=1)

axis_x = []
axis_y = []

for k in range(21,101,5):
    print('[k=%s]\t'%k, '==', end='')
    clf = KMeans(n_clusters = k)
    print('==', end='')
    clf.fit(data_used)
    print('==', end='')
    axis_x.append(k)
    print('==', end='')
    axis_y.append(silhouette_score(data_used, clf.labels_))
    print('== -> Done.')

print('----------')
for n, i in enumerate(axis_x):
    print('k=%s  -> '%i, '%s'%axis_y[n])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(axis_x, axis_y, 'm^--')
plt.title('Sihouette Score')
plt.xlabel('K')
plt.ylabel('Score')
```
