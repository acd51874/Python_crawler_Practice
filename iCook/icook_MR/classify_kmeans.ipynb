```python
import pandas as pd
import os

path = r'./res_analyze_used'

df_origin = pd.read_csv(r'%s/iCook_ingredient_res_analyze_used_v5.csv'%(path), engine='python', encoding='utf-8')
df = df_origin
df_origin
```

```python
'''

Compute the amount of all ingredient
Map-Reduce

'''

ingredient_row_data = df['Ingredients']
ingredient_row_data
```

```python
# Map-Reduce
each_ingredient_amount_dict = {}

for num, each_row in enumerate(ingredient_row_data):
    try:
        for each_ingredient in each_row.split(';'):
            if each_ingredient in each_ingredient_amount_dict:
                each_ingredient_amount_dict[each_ingredient] += 1
            else:
                each_ingredient_amount_dict[each_ingredient] = 1
    except AttributeError:
        print(num, 'AttributeError')
        continue

len(each_ingredient_amount_dict)
```

```python
adv_each_ingredient_amount_dict = {}
for i in each_ingredient_amount_dict:
    if each_ingredient_amount_dict[i] > 500:
        adv_each_ingredient_amount_dict[i] = each_ingredient_amount_dict[i]
len(adv_each_ingredient_amount_dict)
```
