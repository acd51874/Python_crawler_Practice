```python
def translate_ch_to_en(ingredientStr):
    return translator.translate(ingredientStr).text

import re
str = "hello,world!!%[545]你好234世界。。。"
str2 = "冷凍覆盆子;COLD;STONE冰淇淋;無鹽奶油;糖;低筋麵粉;無鹽奶油;牛奶;水;鹽;中筋"
str = re.sub("[A-Za-z0-9\!\%\[\]\,\。]", "", str)
str2 = re.sub("[A-Za-z0-9\!\%\[\]\,\。]", "", str2)
print(str)
print(str2)
```

```python
df = df_origin
errorList = []

for num, i in enumerate(df_origin['Ingredients']):
    print(num, i)
    tmpStr = re.sub("[A-Za-z0-9\!\%\[\]\,\。]", "", i)
    try:
        df['Ingredients'][num] = translate_ch_to_en(tmpStr)
    except:
        print(num, "Error")
        errorList.append(num)
    print(">>", df['Ingredients'][num])
    print("--------------------")
    time.sleep(0.3)

print("\nAll done.")
```
